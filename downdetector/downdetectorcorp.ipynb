```python
# Max Gadeberg
# Aug 29, 2024

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import matplotlib.pyplot as plt

# Define the list of URLs and their specific check elements
urls = {
    'ipa1.it.example.net': {
        'xpath': '//div[@class="col-sm-3 col-md-3 control-label"]/label[@for="username1"]',
        'down_message': 'ðŸ”´ Site is down - unable to find username field at '
    },
    'example.example.net': {
        'xpath': '//div[@data-test="form-control"]//label[@id="loginName-label"]',
        'down_message': 'ðŸ”´ Site is down - unable to find username field at '
    },
    'gitlab.example.net': {
        'xpath': '//div[@class="form-group"]//label[@for="ldapmain_username"]',
        'down_message': 'ðŸ”´ Site is down - unable to find username field at '
    },
    'eng-gitlab.example.net': {
        'xpath': '//div[@class="form-group"]//label[@for="ldapmain_username"]',
        'down_message': 'ðŸ”´ Site is down - unable to find username field at '
    },
    'gitlab.it.example.net': {
        'xpath': '//div[@class="form-group"]//label[@for="ldapmain_username"]',
        'down_message': 'ðŸ”´ Site is down - unable to find username field at '
    },
    'www.example.net': {
        'xpath': '//p[@class="p1"]',
        'down_message': 'ðŸ”´ Site is down - unable to find expected text at '
    }
}

# Set up Selenium WebDriver for Firefox
driver = webdriver.Firefox()

# Dictionary to hold status data
status_data = {}

# Check each URL
for domain, check in urls.items():
    url = f"http://{domain}"
    driver.get(url)
    
    try:
        # Wait for the element to be present and visible
        element = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, check['xpath']))
        )
        if element:
            status = f"ðŸŸ¢ Site is up - found expected element at "
            status_data[url] = status
        else:
            status = check['down_message']
            status_data[url] = status
    except Exception as e:
        status = check['down_message']
        status_data[url] = status

    print(f"{status}{url}")

driver.quit()
```
